# Load and Run Saved CNN Image Classification Models

```python
from tensorflow.keras.models import load_model
import tensorflow as tf
import os
from matplotlib import pyplot as plt
import numpy as np
import cv2

gpus = tf.config.experimental.list_physical_devices("GPU")

for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)

newModel = load_model(os.path.join('models', 'happySadModel4.h5'))
print(newModel)

image = cv2.imread(os.path.join('testing','sadTest.jpg'))

resize = tf.image.resize(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), (256,256))
plt.imshow(resize.numpy().astype(int))
plt.show()

yhatnew = newModel.predict(np.expand_dims(resize/255, 0))

if yhatnew > 0.5:
    print(f'Predicted image is Sad')
else:
    print(f'Predicted image is Happy')
```
